```python
df["SeniorCitizen"]=df["SeniorCitizen"].replace({1:'yes',0:'no'})
```

```python
ax=sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0]) # for counts exact number we want to know
```

```python
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage Of Churned Custmers")
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x="gender",data=df,hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn by Gender")
```

```python
ax=sns.countplot(x="SeniorCitizen",data=df)
ax.bar_label(ax.containers[0])
plt.title("Churn By Gender")
```

```python
# Step 1: Calculate counts
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Step 2: Convert to percentages
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Step 3: Plot
ax = percentages.plot(kind='bar', stacked=True, figsize=(5, 5), colormap='Set2')

# Step 4: Add % labels on each segment
for i, row in enumerate(percentages.values):
    cum_height = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, cum_height + val / 2, f'{val:.1f}%', ha='center', va='center', fontsize=10)
            cum_height += val

# Step 5: Formatting
plt.title("Churn Rate by Senior Citizen Status")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout()
plt.show()
```

```python


# Your selected columns
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Grid: 3 plots per row
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols

plt.figure(figsize=(18, 5 * n_rows))  # Adjust height per row

# Loop through and plot
for i, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, i)
    ax= sns.countplot(data=df, x=col, hue='Churn', order=df[col].value_counts().index, palette='Set2')

    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    plt.title(f'{col}', fontsize=13)
    plt.xticks(rotation=45)
    plt.xlabel("")
    plt.ylabel("Count")

# Main title for all subplots
plt.tight_layout()
plt.suptitle("Count Plots for Telecom Service Features", fontsize=20, y=1.02)
plt.show()
```
